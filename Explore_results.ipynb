# Running the "Main_script.R"
This repository consists of a collection of scripts that can be used to process count tables and create exploratory figures.

If you'd like to know exactly how that is being done, explore the "Main_script.R" script in the root directory of this repository. Briefly, it does the following:
* Points to the location of relevant files containig the microbiome and resistome results
* Lists the metadata variables to use in our data analysis
* Runs additional scripts in the "scripts" directory:
  * To read-in data and create normalized counts
    * scripts/metagenomeSeq_qiime2.R
    * scripts/metagenomeSeq_megares.R
  * To print out aggregated count matrices and exploratory figures:
    * scripts/print_AMR_figures.R
    * scripts/print_microbiome_figures.R
  * To run differential abundance testing using a zero-inflated Gaussian Model (ZIG)
    * scripts/print_AMR_ZIG_results.R
    * scripts/print_microbiome_ZIG_results.R

```r
# You'll notice that no "sparse" features were removed from either the microbiome or the resistome.
# The package, MetagenomeSeq, outputs a warning that the Default quantile value is being used for normalization.
# Lastly, you'll see output from NMDS ordination as the scripts recreates exploratory figures in the "graphs" directory.
source("Main_script.R")
```

# Code for statistical comparisons in manuscript:
Below you'll find the code used to make statistical comparisons of:
* sequencing results with linear models
* mapping results with linear models
* diversity indices with the Wilcoxon test
* Ordination clustering with ANOSIM

For the differential abundance tests, results can be found in the "stats" directory. The code for replicating this analysis is part of the "print_microbiome_ZIG_results.R" script with further details in "meg_utility_functions.R".

## Linear models comparing sequencing and mapping results

```r
#
##
### GLM for read counts by Treatment, Dilution, Packaging
##
#

summary(glm(metadata$Raw_paired_reads ~ as.factor(metadata$Treatment)))
summary(glm(metadata$Raw_paired_reads ~ as.factor(metadata$Dilution)))
summary(glm(metadata$Raw_paired_reads ~ as.factor(metadata$Packaging)))

summary(glm(metadata$QC_filtered_reads ~ as.factor(metadata$Treatment)))
summary(glm(metadata$QC_filtered_reads ~ as.factor(metadata$Dilution)))
summary(glm(metadata$QC_filtered_reads ~ as.factor(metadata$Packaging)))

summary(glm(metadata$nonhost_filtered_reads ~ as.factor(metadata$Treatment)))
summary(glm(metadata$nonhost_filtered_reads ~ as.factor(metadata$Dilution)))
summary(glm(metadata$nonhost_filtered_reads ~ as.factor(metadata$Packaging)))

summary(glm(metadata$deduped_SNP_confirmed_counts ~ as.factor(metadata$Treatment)))
summary(glm(metadata$deduped_SNP_confirmed_counts ~ as.factor(metadata$Dilution)))
summary(glm(metadata$deduped_SNP_confirmed_counts ~ as.factor(metadata$Packaging)))

## GLM microbome
summary(glm(microbiome_metadata$Raw_paired_reads ~ as.factor(microbiome_metadata$Treatment)))
summary(glm(microbiome_metadata$Raw_paired_reads ~ as.factor(microbiome_metadata$Packaging)))

summary(glm(microbiome_metadata$QC_filtered_reads ~ as.factor(microbiome_metadata$Treatment)))
summary(glm(microbiome_metadata$QC_filtered_reads ~ as.factor(microbiome_metadata$Packaging)))

summary(glm(microbiome_metadata$nonhost_filtered_reads ~ as.factor(microbiome_metadata$Treatment)))
summary(glm(microbiome_metadata$nonhost_filtered_reads ~ as.factor(microbiome_metadata$Dilution)))
```

## Resistome statistical comparisons

```r
## Wilcoxon test of diversity indices by Treatment and Dilution
#
# Test by treatment, Class and Mech
wilcox.test(metadata$AMR_class_Richness ~ metadata$Treatment)
wilcox.test(metadata$AMR_class_Shannon ~ metadata$Treatment)
wilcox.test(metadata$AMR_class_Richness ~ metadata$Treatment)
wilcox.test(metadata$AMR_class_Shannon ~ metadata$Treatment)
# Test by Dilution, Class and Mech
wilcox.test(metadata$AMR_class_Richness ~ metadata$Dilution)
wilcox.test(metadata$AMR_class_Shannon ~ metadata$Dilution)
wilcox.test(metadata$AMR_class_Richness ~ metadata$Dilution)
wilcox.test(metadata$AMR_class_Shannon ~ metadata$Dilution)

## ANOSIM clustering at the Class and Mechanism level
#
# Dilution
anosim(metaMDS_AMR_class$points,pData(AMR_analytic_data[[1]])$Dilution, distance='euclidean')
anosim(metaMDS_AMR_mech$points,pData(AMR_analytic_data[[2]])$Dilution, distance='euclidean')
# Treatment
anosim(metaMDS_AMR_class$points,pData(AMR_analytic_data[[1]])$Treatment, distance='euclidean')
anosim(metaMDS_AMR_mech$points,pData(AMR_analytic_data[[2]])$Treatment, distance='euclidean')
# Packaging
anosim(metaMDS_AMR_class$points,pData(AMR_analytic_data[[1]])$Packaging, distance='euclidean')
anosim(metaMDS_AMR_mech$points,pData(AMR_analytic_data[[2]])$Packaging, distance='euclidean')
# Blinded Store
anosim(metaMDS_AMR_class$points,pData(AMR_analytic_data[[1]])$Blinded_Store, distance='euclidean')
anosim(metaMDS_AMR_mech$points,pData(AMR_analytic_data[[2]])$Blinded_Store, distance='euclidean')
```
